```python


import pandas as pd
import matplotlib.pyplot as plt

# Read the CSV file with the specified columns
project = pd.read_csv(r"/content/drive/MyDrive/DATA/Most Streamed Spotify Songs 2024.csv",
                      encoding='latin-1',
                      usecols=["Track Score", "Spotify Streams", "Pandora Streams", "Soundcloud Streams"])

# Remove commas from the values and convert columns to numeric values
# not needed project["Track Score"] = pd.to_numeric(project["Track Score"].str.replace(',', ''), errors='coerce')
project["Spotify Streams"] = pd.to_numeric(project["Spotify Streams"].str.replace(',', ''), errors='coerce')
project["Pandora Streams"] = pd.to_numeric(project["Pandora Streams"].str.replace(',', ''), errors='coerce')
project["Soundcloud Streams"] = pd.to_numeric(project["Soundcloud Streams"].str.replace(',', ''), errors='coerce')

# Drop rows where "Spotify Streams" is less than 1,000,000,000
# project = project[project["Spotify Streams"] >= 1000000000]

# Drop duplicates
project = project.drop_duplicates()

# Convert stream values to tens of millions
project["Spotify Streams"] /= 10**7
project["Pandora Streams"] /= 10**7
project["Soundcloud Streams"] /= 10**7

# Plotting
plt.figure(figsize=(10, 6))

# Scatter plot for "Spotify Streams"
plt.scatter(project["Track Score"], project["Spotify Streams"], label="Spotify Streams", alpha=0.5)

# Scatter plot for "Pandora Streams"
plt.scatter(project["Track Score"], project["Pandora Streams"], label="Pandora Streams", alpha=0.5)

# Scatter plot for "Soundcloud Streams"
plt.scatter(project["Track Score"], project["Soundcloud Streams"], label="Soundcloud Streams", alpha=0.5)

# Adding labels and title
plt.xlabel("Track Score")
plt.ylabel("Streams (in tens of millions)")
plt.title("Track Score vs. Streams")
plt.legend()

# Display the plot
plt.show()
```
